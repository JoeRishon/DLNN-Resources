The steps that we are going to follow:

Load Data.  
Define Keras Model.  
Compile Keras Model.  
Fit Keras Model.  
Evaluate Keras Model.  
Make Predictions  
Save Model   

Import the following libraries

```
import keras
import numpy as np
import pandas as pd
```

## 1.Load Text

```
df=pd.read_csv('/content/pima.csv')
df.head()
```

```
df
```

```
X = df.drop(columns=['variable'])
Y = df[['variable']]
```

```
X
```

```
Y
```

## 2.Define Keras Model

```
model = keras.models.Sequential()

model.add(keras.layers.Dense(12, activation='relu', input_shape=(8,)))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

## 3. Compile Keras Model

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 4. Fit Keras Model

```
model.fit(X, Y, epochs=150, batch_size=10)
```

## 5. Evaluate Keras Model

```
_, accuracy = model.evaluate(X, Y)
print('Accuracy: %.2f' % (accuracy*100))
```

## 6. Make Predictions

```
predictions = model.predict_classes(X)
# summarize the first 5 cases
Xlist=X.values.tolist()
Ylist=Y.values.tolist()
for i in range(5):
	print('%s => %d (expected %d)' % (Xlist[i], predictions[i][0], Ylist[i][0]))
```

## 7. Saving model for later use

```
model.save('saved_model.h5')
```

```
old_model = keras.models.load_model('saved_model.h5')
```

```
_, accuracy = old_model.evaluate(X, Y)
print('Accuracy: %.2f' % (accuracy*100))
```
